````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fredapi import Fred
from statsmodels.tsa.arima.model import ARIMA
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from math import sqrt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from prophet import Prophet

# Initialize the FRED API with your API key
fred = Fred(api_key='b43431d564515744c751e7a57be340ac')

# Fetch CPI and Federal Funds Rate data from FRED
cpi_data = fred.get_series('CPIAUCSL', observation_start='1950-01-01')
ffr_data = fred.get_series('FEDFUNDS', observation_start='1950-01-01')

# Convert the series to DataFrames and rename columns
df_cpi = pd.DataFrame(cpi_data, columns=['CPI'])
df_ffr = pd.DataFrame(ffr_data, columns=['Federal_Funds_Rate'])
df_cpi.index = pd.to_datetime(df_cpi.index)
df_ffr.index = pd.to_datetime(df_ffr.index)

# Merge the two dataframes
df = df_cpi.join(df_ffr, how='inner')

# Function to plot forecasts
def plot_forecast(df, forecast, forecast_dates, title, ylabel):
    plt.figure(figsize=(10, 6))
    plt.plot(df.index, df[ylabel], label='Historical Data', marker='o')
    plt.plot(forecast_dates, forecast, 'r*', label='Forecasted Data', markersize=10)
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel(ylabel)
    plt.legend()
    plt.tight_layout()
    plt.show()

# ARIMA Forecasting for CPI
model_arima_cpi = ARIMA(df['CPI'], order=(9, 1, 0))  # Define the ARIMA model with order (9, 1, 0)
model_fit_arima_cpi = model_arima_cpi.fit()  # Fit the model to the CPI data
forecast_arima_cpi = model_fit_arima_cpi.forecast(steps=12)  # Forecast the next 12 months
forecast_dates_arima_cpi = pd.date_range(start=df.index[-1], periods=13, freq='M')[1:]  # Define the forecast dates

# ARIMA Forecasting for Federal Funds Rate
model_arima_ffr = ARIMA(df['Federal_Funds_Rate'], order=(9, 1, 0))  # Define the ARIMA model with order (9, 1, 0)
model_fit_arima_ffr = model_arima_ffr.fit()  # Fit the model to the Federal Funds Rate data
forecast_arima_ffr = model_fit_arima_ffr.forecast(steps=12)  # Forecast the next 12 months
forecast_dates_arima_ffr = pd.date_range(start=df.index[-1], periods=13, freq='M')[1:]  # Define the forecast dates

# LSTM Forecasting

# Normalize the data to the range (0, 1)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_values = scaler.fit_transform(df.values)

# Function to convert time series data to supervised learning format
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    df = pd.DataFrame(data)
    cols, names = [], []
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [(f'var{j+1}(t-{i})') for j in range(data.shape[1])]
    for i in range(n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [(f'var{j+1}(t)') for j in range(data.shape[1])]
        else:
            names += [(f'var{j+1}(t+{i})') for j in range(data.shape[1])]
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    if dropnan:
        agg.dropna(inplace=True)
    return agg

# Convert the normalized series to a supervised learning format
reframed = series_to_supervised(scaled_values, 1, 1)
values = reframed.values

# Split the data into training and testing sets
n_train_time = int(len(values) * 0.9)
train = values[:n_train_time, :]
test = values[n_train_time:, :]

# Separate the inputs and outputs for training and testing sets
train_X, train_y = train[:, :-2], train[:, -2:]
test_X, test_y = test[:, :-2], test[:, -2:]

# Reshape the inputs to be 3D [samples, timesteps, features] for LSTM
train_X = train_X.reshape((train_X.shape[0], 1, train_X.shape[1]))
test_X = test_X.reshape((test_X.shape[0], 1, test_X.shape[1]))

# Build the LSTM model
model_lstm = Sequential()
model_lstm.add(LSTM(50, input_shape=(train_X.shape[1], train_X.shape[2])))
model_lstm.add(Dense(2))  # Two outputs: CPI and Federal Funds Rate
model_lstm.compile(optimizer='adam', loss='mae')

# Train the LSTM model
model_lstm.fit(train_X, train_y, epochs=500, batch_size=72, validation_data=(test_X, test_y), verbose=2, shuffle=False)

# Forecast using the LSTM model
input_data = np.array([df.values[-1, :]])
input_data = scaler.transform(input_data)
input_data = input_data.reshape((input_data.shape[0], 1, input_data.shape[1]))

forecast_lstm_cpi = []
forecast_lstm_ffr = []

for i in range(12):
    forecast = model_lstm.predict(input_data)
    forecast_lstm_cpi.append(forecast[0, 0])
    forecast_lstm_ffr.append(forecast[0, 1])
    input_data = np.append(input_data[:, 1:, :], forecast.reshape(1, 1, -1), axis=1)

# Inverse transform the forecasted values to original scale
forecast_lstm_cpi = np.array(forecast_lstm_cpi).reshape(-1, 1)
forecast_lstm_ffr = np.array(forecast_lstm_ffr).reshape(-1, 1)
forecast_lstm_cpi = scaler.inverse_transform(np.concatenate((forecast_lstm_cpi, np.zeros_like(forecast_lstm_cpi)), axis=1))[:, 0]
forecast_lstm_ffr = scaler.inverse_transform(np.concatenate((np.zeros_like(forecast_lstm_ffr), forecast_lstm_ffr), axis=1))[:, 1]
forecast_dates_lstm = pd.date_range(start=df.index[-1], periods=13, freq='M')[1:]

# Prophet Forecasting for CPI
df_cpi_prophet = df_cpi.reset_index().rename(columns={'index': 'ds', 'CPI': 'y'})
model_prophet_cpi = Prophet()
model_prophet_cpi.fit(df_cpi_prophet)
future_prophet_cpi = model_prophet_cpi.make_future_dataframe(periods=12, freq='M')
forecast_prophet_cpi = model_prophet_cpi.predict(future_prophet_cpi)
forecast_dates_prophet_cpi = future_prophet_cpi['ds'].tail(12).values
forecast_prophet_cpi = forecast_prophet_cpi['yhat'].tail(12).values

# Prophet Forecasting for Federal Funds Rate
df_ffr_prophet = df_ffr.reset_index().rename(columns={'index': 'ds', 'Federal_Funds_Rate': 'y'})
model_prophet_ffr = Prophet()
model_prophet_ffr.fit(df_ffr_prophet)
future_prophet_ffr = model_prophet_ffr.make_future_dataframe(periods=12, freq='M')
forecast_prophet_ffr = model_prophet_ffr.predict(future_prophet_ffr)
forecast_dates_prophet_ffr = future_prophet_ffr['ds'].tail(12).values
forecast_prophet_ffr = forecast_prophet_ffr['yhat'].tail(12).values

# Plotting the forecasts for each model
plot_forecast(df, forecast_arima_cpi, forecast_dates_arima_cpi, 'ARIMA Forecast for CPI', 'CPI')
plot_forecast(df, forecast_arima_ffr, forecast_dates_arima_ffr, 'ARIMA Forecast for Federal Funds Rate', 'Federal_Funds_Rate')
plot_forecast(df, forecast_lstm_cpi, forecast_dates_lstm, 'LSTM Forecast for CPI', 'CPI')
plot_forecast(df, forecast_lstm_ffr, forecast_dates_lstm, 'LSTM Forecast for Federal Funds Rate', 'Federal_Funds_Rate')

# Plotting the forecast for Prophet model separately
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['CPI'], label='Historical CPI', marker='o')
plt.plot(forecast_dates_prophet_cpi, forecast_prophet_cpi, 'r*', label='Forecasted CPI', markersize=10)
plt.title('Forecasted CPI using Prophet Model')
plt.xlabel('Date')
plt.ylabel('CPI')
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Federal_Funds_Rate'], label='Historical Federal Funds Rate', marker='o')
plt.plot(forecast_dates_prophet_ffr, forecast_prophet_ffr, 'r*', label='Forecasted Federal Funds Rate', markersize=10)
plt.title('Forecasted Federal Funds Rate using Prophet Model')
plt.xlabel('Date')
plt.ylabel('Federal_Funds_Rate')
plt.legend()
plt.tight_layout()
plt.show()

# Printing the forecasted values and percentage changes
models = ['ARIMA', 'LSTM', 'Prophet']
forecasts_cpi = [forecast_arima_cpi, forecast_lstm_cpi, forecast_prophet_cpi]
forecasts_ffr = [forecast_arima_ffr, forecast_lstm_ffr, forecast_prophet_ffr]
forecast_dates_cpi = [forecast_dates_arima_cpi, forecast_dates_lstm, forecast_dates_prophet_cpi]
forecast_dates_ffr = [forecast_dates_arima_ffr, forecast_dates_lstm, forecast_dates_prophet_ffr]

for model_name, forecast, dates in zip(models, forecasts_cpi, forecast_dates_cpi):
    print(f'\n{model_name} Forecast for CPI:')
    last_observed_cpi = df['CPI'].iloc[-1]
    for date, value in zip(dates, forecast):
        date = pd.Timestamp(date)
        percentage_change = ((value - last_observed_cpi) / last_observed_cpi) * 100
        print(f"Forecasted CPI for {date.year}-{date.month}: {value:.2f} ({percentage_change:.2f}% from last observed CPI)")

for model_name, forecast, dates in zip(models, forecasts_ffr, forecast_dates_ffr):
    print(f'\n{model_name} Forecast for Federal Funds Rate:')
    last_observed_ffr = df['Federal_Funds_Rate'].iloc[-1]
    for date, value in zip(dates, forecast):
        date = pd.Timestamp(date)
        percentage_change = ((value - last_observed_ffr) / last_observed_ffr) * 100
        print(f"Forecasted Federal Funds Rate for {date.year}-{date.month}: {value:.2f} ({percentage_change:.2f}% from last observed Federal Funds Rate)")

# Comparative Analysis with RMSE
# Split the data into training and testing sets for RMSE calculation
train_data_cpi = df['CPI'].iloc[:-12]  # Use all but the last 12 months for training
test_data_cpi = df['CPI'].iloc[-12:]  # Use the last 12 months for testing

train_data_ffr = df['Federal_Funds_Rate'].iloc[:-12]  # Use all but the last 12 months for training
test_data_ffr = df['Federal_Funds_Rate'].iloc[-12:]  # Use the last 12 months for testing

# ARIMA model RMSE for CPI
model_arima_cpi = ARIMA(train_data_cpi, order=(9, 1, 0))
model_fit_arima_cpi = model_arima_cpi.fit()
forecast_arima_test_cpi = model_fit_arima_cpi.forecast(steps=len(test_data_cpi))
rmse_arima_cpi = sqrt(mean_squared_error(test_data_cpi, forecast_arima_test_cpi))

# ARIMA model RMSE for Federal Funds Rate
model_arima_ffr = ARIMA(train_data_ffr, order=(9, 1, 0))
model_fit_arima_ffr = model_arima_ffr.fit()
forecast_arima_test_ffr = model_fit_arima_ffr.forecast(steps=len(test_data_ffr))
rmse_arima_ffr = sqrt(mean_squared_error(test_data_ffr, forecast_arima_test_ffr))

# LSTM model RMSE for CPI
test_X_lstm_cpi = test_X[:, :, :2]  # Correct the input shape for LSTM
yhat_lstm_cpi = model_lstm.predict(test_X_lstm_cpi)
test_X_lstm_cpi = test_X_lstm_cpi.reshape((test_X_lstm_cpi.shape[0], test_X_lstm_cpi.shape[2]))
inv_yhat_lstm_cpi = np.concatenate((yhat_lstm_cpi[:, :1], test_X_lstm_cpi[:, 1:]), axis=1)
inv_yhat_lstm_cpi = scaler.inverse_transform(inv_yhat_lstm_cpi)
inv_yhat_lstm_cpi = inv_yhat_lstm_cpi[:, 0]
inv_y_lstm_cpi = np.concatenate((test_y[:, :1].reshape(-1, 1), test_X_lstm_cpi[:, 1:]), axis=1)
inv_y_lstm_cpi = scaler.inverse_transform(inv_y_lstm_cpi)
inv_y_lstm_cpi = inv_y_lstm_cpi[:, 0]
rmse_lstm_cpi = sqrt(mean_squared_error(inv_y_lstm_cpi, inv_yhat_lstm_cpi))

# LSTM model RMSE for Federal Funds Rate
test_X_lstm_ffr = test_X[:, :, :2]  # Correct the input shape for LSTM
yhat_lstm_ffr = model_lstm.predict(test_X_lstm_ffr)
test_X_lstm_ffr = test_X_lstm_ffr.reshape((test_X_lstm_ffr.shape[0], test_X_lstm_ffr.shape[2]))
inv_yhat_lstm_ffr = np.concatenate((test_X_lstm_ffr[:, :1], yhat_lstm_ffr[:, 1:]), axis=1)
inv_yhat_lstm_ffr = scaler.inverse_transform(inv_yhat_lstm_ffr)
inv_yhat_lstm_ffr = inv_yhat_lstm_ffr[:, 1]
inv_y_lstm_ffr = np.concatenate((test_X_lstm_ffr[:, :1], test_y[:, 1:].reshape(-1, 1)), axis=1)
inv_y_lstm_ffr = scaler.inverse_transform(inv_y_lstm_ffr)
inv_y_lstm_ffr = inv_y_lstm_ffr[:, 1]
rmse_lstm_ffr = sqrt(mean_squared_error(inv_y_lstm_ffr, inv_yhat_lstm_ffr))

# Prophet model RMSE for CPI
df_test_prophet_cpi = df_cpi_prophet.iloc[-12:]
forecast_prophet_test_cpi = model_prophet_cpi.predict(df_test_prophet_cpi)
forecast_prophet_test_cpi = forecast_prophet_test_cpi['yhat'].values
rmse_prophet_cpi = sqrt(mean_squared_error(df_test_prophet_cpi['y'], forecast_prophet_test_cpi))

# Prophet model RMSE for Federal Funds Rate
df_test_prophet_ffr = df_ffr_prophet.iloc[-12:]
forecast_prophet_test_ffr = model_prophet_ffr.predict(df_test_prophet_ffr)
forecast_prophet_test_ffr = forecast_prophet_test_ffr['yhat'].values
rmse_prophet_ffr = sqrt(mean_squared_error(df_test_prophet_ffr['y'], forecast_prophet_test_ffr))

print(f'\nRMSE Comparison for CPI:')
print(f'ARIMA: {rmse_arima_cpi:.2f}')
print(f'LSTM: {rmse_lstm_cpi:.2f}')
print(f'Prophet: {rmse_prophet_cpi:.2f}')

print(f'\nRMSE Comparison for Federal Funds Rate:')
print(f'ARIMA: {rmse_arima_ffr:.2f}')
print(f'LSTM: {rmse_lstm_ffr:.2f}')
print(f'Prophet: {rmse_prophet_ffr:.2f}')
````

````python
from textblob import TextBlob

# Function to generate interpretations using sentiment analysis
def generate_interpretation(forecast_values, last_observed_values, label):
    interpretations = []
    for value, last_observed_value in zip(forecast_values, last_observed_values):
        sentiment = TextBlob(f"The forecasted {label} is {value:.2f}").sentiment.polarity
        if sentiment > 0:
            interpretation = "Positive trend: The forecast suggests an increase in the " + label + "."
        elif sentiment < 0:
            interpretation = "Negative trend: The forecast suggests a decrease in the " + label + "."
        else:
            interpretation = "Neutral: The forecast suggests stability in the " + label + "."
        interpretations.append(interpretation)
    return interpretations

# Generate interpretations for CPI forecast
cpi_interpretations = generate_interpretation(forecast_arima_cpi, df['CPI'].iloc[-1:], "CPI")

# Generate interpretations for Federal Funds Rate forecast
ffr_interpretations = generate_interpretation(forecast_arima_ffr, df['Federal_Funds_Rate'].iloc[-1:], "Federal Funds Rate")

# Print interpretations
print("\nInterpretations for CPI Forecast:")
for i, interpretation in enumerate(cpi_interpretations):
    print(f"\nForecast {i+1}:")
    print(interpretation)

print("\nInterpretations for Federal Funds Rate Forecast:")
for i, interpretation in enumerate(ffr_interpretations):
    print(f"\nForecast {i+1}:")
    print(interpretation)
````
